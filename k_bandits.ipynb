# k- Armed Bandit Problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D=np.random.normal(0,1,10)
simulations=2000
steps=5000
```

```python
print("Reward from each bandit is as follows:")
print()
for i in range(1,11):
    print(f"Bandit({i})~N({D[i-1]},1)" )
```

```python
for eps in [0.0,0.01,0.05,0.1]:
    rewardsum=np.array([0 for j in range(steps)])
    for i in range(simulations):
        reward=np.array([0 for j in range(steps)])
        Q=[0 for i in range(10)]
        N=[0 for i in range(10)]
        for j in range(steps):
            if np.random.uniform(0,1)<eps:
                A=np.random.randint(0,10)
                
            else:
                max_indices = np.where(Q == np.max(Q))[0]
                A=np.random.choice(max_indices)
            reward[j]=np.random.normal(D[A],1,1)
            N[A]+=1
            Q[A]+=(reward[j]-Q[A])/N[A]
        rewardsum+=reward
    rewardsum=rewardsum/simulations
    plt.plot(rewardsum,label=str(eps))
    print(eps)
plt.legend()
plt.show()
        
           
        
        
    
```

Above is a plot of average reward at each step. average is calculated over 2000 simulations for each epsilon in {0,0.01,0,05,0.1}.

As can be seen average reward of pure greedy approach is lesser than the epsilon greedy.
Epsilon=0.05 is performing much better in terms of average reward than pure greedy in this example.
